```python
def get_top_n_groups(n):
    with open(PATH, 'r') as file: 
        return json.load(file)[:n]
```

```python
def get_group_members(id):
    params = dict(access_token=ACCESS_TOKEN, v=VERSION, group_id=id, fields='sex, bdate')
    response = requests.get('https://api.vk.com/method/groups.getMembers', params).json()
    if response['response']['count'] > 1000:
        users = get_via_execute(params, 'groups.getMembers', response['response']['count'], id)
    else:
        users = response['response']['items']
    return users
```

```python
def get_data(data):
    age_list = []
    sex_list = []
    for element in data:
        sex_list.append(element['sex'])
        if not element.get('bdate') or element['bdate'].count('.') != 2:
            continue
        age_list.append(get_age(element['bdate']))
    return sex_list, age_list
```

```python
def get_age(bdate):
    try:
        bdate = datetime.date(*list(map(int, bdate.split('.')[::-1])))
    except ValueError:
        return 0
    years = (TODAY - bdate).days // 365
    return years
```

```python
def print_sex_bar(data_list, name):
    data = [data_list.count(2), data_list.count(1)]
    index = ['Male','Female']
    pd.DataFrame({'Qauntity': [*data]}, index = [*index]).plot.bar(title=name)
```

```python
def print_age_bar(data_list, name):
    data = [
        len([x for x in data_list if 10 <= x <= 15]),
        len([x for x in data_list if 16 <= x <= 21]),
        len([x for x in data_list if 22 <= x <= 30]),
        len([x for x in data_list if 31 <= x <= 40]),
        len([x for x in data_list if 41 <= x <= 80]),
        len([x for x in data_list if 81 <= x])
    ]
    index = ['10-15', '16-21', '22-30', '31-40', '41-80', '>81']
    pd.DataFrame({'Qauntity': [*data]}, index = [*index]).plot.bar(title=name)
```

```python
def get_via_execute(dict_params, method, number, group_id):
    number_of_query = [25000 for x in range(number // 25000)]
    number_of_query.append(number % 25000)
    dict_params = dict(access_token=ACCESS_TOKEN, v=VERSION, group_id=id, fields='sex, bdate')
    users = []
    offset = 0
    for i in number_of_query:
        number_of_execute = round((i + 499) / 1000)
        execute = 'var members = [];var i = {i};var offset = {offset};while (i > 0){open_bracket}members.push(API.{method}({open_bracket}"group_id": {group_id}, "v": {v}, "fields": "sex, bdate", "count": "1000", "offset": offset{close_bracket}).items);offset = offset + 1000;i = i - 1;{close_bracket}return members;'.format(i=number_of_execute, open_bracket='{', method=method, offset=offset,
                                           group_id=group_id, v=dict_params['v'], close_bracket='}')
                  
        offset += i
        dict_params['code'] = execute
        response = requests.get('https://api.vk.com/method/execute', dict_params).json()
        for element in response['response']:
            users.extend(element)
    return users
```
